# SubTask 4.1
##### Colleague suggests that non price aspects can be modelled using two latent variables: 
* Quality (condition and grade)
* Size (3 size measures)

These are the latent variables we are looking for, and so, we should check for correlatation coefficients

## Discovering Underlying Factors

```python
df.corr(method='pearson')
```

##### Quality (capturing condition and grade) 
* No meaningful correlation between condition and grade

##### Size (capturing the three size measures)
* sqft above and sqft_living15 have a reasonably high correlation, as to be expected as sqft above is defined as Square footage of house apart from basement, which would naturally include the living room area. 
* Basement has no meaningful correlation with remaining two size variables, to be expected from what we know these features to represent.
* In fact, grade has a high correlation to sqft_living and sqft_above... as well as the target variable price

**Doesn't support the hypothesis of proposed latent variables**

```python
factorA = FactorAnalysis(n_components = 2)
facA = factorA.fit_transform(X)
pdf = pd.DataFrame(data = facA, columns =['principal comp0','principal comp1'])
pdf.head()
```

```python
compMatrix = pd.DataFrame(data = factorA.components_, columns = features[-5:])
compMatrix
```

## Justification & Analysis
* Doesn't really fit the expectations outlined for proposed Quality and Size (not interpretable).
For **Quality** I would expect an obvious correlation between condition and grade and vice-versa for size, whereas in actual fact the features have strong impacts over both components.
* If this data were as easily as interpretable (as seen in car example), then an obvious linear combination of components for both size and quality would exist. In the car example there are clear 0 values for features of length in the price component that clearly indentify/distinguish the latent variables from oneanother; the actual solution is nonidentifiable.

## To improve interpretability: Apply Constraints
* Othonormal
* upper diagonal
* sparsity

```python
fig = plt.figure(figsize=(11,11))
ax = fig.add_subplot(111)
ax.scatter(pdf['principal comp0'], pdf['principal comp1'] )
ax.set_ylabel("PC2")
ax.set_xlabel("PC1")
plt.show()
```
